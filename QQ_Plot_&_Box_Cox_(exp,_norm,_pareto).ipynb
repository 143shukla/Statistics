```python
 # Sample of random variable which follows random distribution
 normal_data = np.random.normal(loc = 20, scale = 10, size = 100)
```

```python
# QQ- PLot
stats.probplot(normal_data, dist = 'norm', plot = plt)
plt.grid()
```

```python
# Increasing the sample size
normal_data = np.random.normal(loc = 20, scale = 10, size = 1000)
stats.probplot(normal_data, dist = 'norm', plot = plt)
plt.grid()
```

```python
exponancial_data = np.random.exponential(scale=2, size=100)
stats.probplot(exponancial_data, dist='norm', plot=plt)
plt.grid()
```

```python
exponancial_data = np.random.exponential(scale=2, size=100)
stats.probplot(exponancial_data, dist='norm', plot=plt)
plt.grid()
```

```python
exponential_data = np.random.exponential(scale=2, size=100)
stats.probplot(exponential_data, dist='norm', plot=plt)
plt.grid()
```

```python
sn.distplot(exponential_data)
plt.grid()
```

```python
# plot CDF(X)
kwargs = {'cumulative': True}
sn.distplot(exponential_data, hist_kws=kwargs, kde_kws=kwargs)
```

```python
sn.kdeplot(exponential_data, cumulative=True)
```

Box-Cox Transform: https://towardsdatascience.com/box-cox-transformation-explained-51d745e34203

```python
transformed_data, l = stats.boxcox(exponential_data)
```

```python
stats.probplot(transformed_data, dist='norm', plot=plt)
plt.grid()
```

```python
# PDF plot of box-cox trandformed data
sn.set()
sn.distplot(transformed_data)

# CDF plot of box-cox trandformed data
kwargs = {'cumulative': True}
sn.distplot(transformed_data, hist_kws=kwargs, kde_kws=kwargs)
```

Pareto Distribution

```python
pareto_data = np.random.pareto(a = 2.0, size = 1000)
sn.distplot(pareto_data)

kwargs = {'cumulative' : True}
sn.distplot(pareto_data, hist_kws=kwargs, kde_kws=kwargs)
plt.grid()
```

```python
# OR
sn.kdeplot(pareto_data)
sn.kdeplot(pareto_data,cumulative=True)
```

```python
# Transforming Preto data using box-cox
pareto_transformed, p = stats.boxcox(pareto_data)

stats.probplot(pareto_transformed, dist = 'norm', plot = plt)
plt.grid()
```

```python
# PDF and CDF of pareto data transformed using box-cox
sn.distplot(pareto_transformed)

kwargs = {'cumulative': True}
sn.distplot(pareto_transformed, hist_kws=kwargs, kde_kws=kwargs)
plt.grid()
```
